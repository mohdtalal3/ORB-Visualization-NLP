```python
import json
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nrclex import NRCLex
with open('conver.json', 'r') as json_file:
    data = json.load(json_file)
analyzer = SentimentIntensityAnalyzer()

def analyze_emotions(conversation,word_freq):
    nrc_result = NRCLex(conversation)
    emotio = nrc_result.affect_frequencies
    emotion_words = nrc_result.affect_dict
    emotion_categories = {
        "fear": [],
        "anger": [],
        "anticip": [],
        "trust": [],
        "surprise": [],
        "positive": [],
        "negative": [],
        "sadness": [],
        "disgust": [],
        "joy": [],
        "anticipation": []
    }
    for word, emotions in emotion_words.items():
        for emotion in emotions:
            for category in emotion_categories.keys():
                if category in emotion:
                    if word in word_freq:
                        frequency = word_freq[word]
                        emotion_categories[category].append((word,frequency))

    
    result = {
        "emotions": emotio,
        "emotion_words": emotion_categories
    }
    
    return result
result_data = []

for item in data:
    sender = item["sender"]
    receiver_list = item["receiverList"]
    
    receiver_results = []
    
    for receiver_item in receiver_list:
        conversation = receiver_item[" conversation"]
        from collections import Counter
        import re
        conversation = receiver_item[" conversation"]
        words = re.findall(r'\w+', conversation.lower())  
        word_freq = Counter(words)
        emotions_result = analyze_emotions(conversation,word_freq)
        
        receiver_result = {
            "receiver": receiver_item["receiver"],
            "conversation_result": emotions_result
        }
        
        receiver_results.append(receiver_result)
    
    sender_result = {
        "sender": sender,
        "receiverList_results": receiver_results
    }
    
    result_data.append(sender_result)
with open('conver1.json', 'w') as json_file:
    json.dump(result_data,json_file, indent=4)
```
